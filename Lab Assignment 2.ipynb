```python
from tclab import TCLab, clock, Historian, Plotter
import time

%matplotlib inline
import matplotlib.pyplot as plt

with TCLab() as lab:
    P1 = 200
    P2 = 100
    h = Historian(lab.sources)
    p = Plotter(h,1800)
    
    def u1(t):
        if (t < 20):
            Q1 = lab.Q1(0)
        elif (t < 620):
            Q1 = lab.Q1(60)
        elif (t < 1020):
            Q1 = lab.Q1(60)
        elif (t >= 1020):
            Q1 = lab.Q1(0)
        
    def u2(t):
        if (t < 620):
            Q2 = lab.Q2(0)
        elif (t <1020):
            Q2 = lab.Q2(60)
        elif (t >= 1020):
            Q2 = lab.Q2(0)
            
    for t in clock(1800):
        u1(t)
        u2(t)
        p.update(t)
            
h.to_csv('Lab2_Data.csv')

```
